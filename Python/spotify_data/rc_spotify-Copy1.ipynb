# Get ALL tracks from playlist

```python
offset = 0

data = [1,1,1]

new = []

while offset < 10:
    for i in data:
        
        new.append(data[i] + 1)
        offset += 1



# loop through playlist to extract song data
user = 'spotify:user:USERID'

pl_1 = 'spotify:user:USER:playlist:ID1'
pl_2 = 'spotify:user:USER:playlist:ID1'

# returns 100 tracks at a time
#pl_1_data = sp.user_playlist_tracks(user, pl_1, offset=100)['items']

data = []
data_2 = []
offset = 0


while offset < 10000:

    playlist = sp.user_playlist_tracks(user, pl_2, offset=offset)['items']

    for item in playlist:

        album_uri = item['track']['album']['uri']
        release_date = item['track']['album']['release_date']
        artist_name = item['track']['artists'][0]['name']
        artist_uri = item['track']['artists'][0]['uri']

        track_name = item['track']['name']
        track_uri = item['track']['uri']
        popularity = item['track']['popularity']

        data_2.append({
            'album_uri': album_uri,
            'release_date': release_date,
            'artist_name': artist_name,
            'artist_uri': artist_uri,
            'track_name': track_name,
            'track_uri': track_uri,
            'popularity':popularity
        })

    offset += 100
    print(len(data_2))
```

```python
df_1 = pd.DataFrame(data, columns=['album_uri', 'release_date', 'artist_name', 'artist_uri',
                                  'track_name', 'track_uri', 'popularity'])

df_1.drop_duplicates(['track_uri'], inplace=True)

df_1.to_csv('df_1.csv', index=False)

df_1.info()
```

```python
df_2 = pd.DataFrame(data_2, columns=['album_uri', 'release_date', 'artist_name', 'artist_uri',
                                  'track_name', 'track_uri', 'popularity'])

df_2.drop_duplicates(['track_uri'], inplace=True)

df_2.to_csv('df_2.csv', index=False)

df_2.info()
```

```python
df_master = pd.concat([df_1, df_2], ignore_index=True)

df_master.to_csv('df_master.csv', index=False)

df_master.info()
```

```python
# set client id, secret and redirect uri
client_id='YOUR_CLIENT_ID'
client_secret='YOUR_CLIENT_SECRET'
redirect_uri='http://your_redirect.com/'

# column names for DF
cols = ['id','uri','analysis_url','track_href','type','duration_ms',
        'acousticness','danceability','energy','instrumentalness','liveness',
        'loudness','speechiness','valence','key','mode','tempo','time_signature']

# new column names for DF
new_cols = ['id', 'track_uri', 'analysis_url', 'track_href', 'type', 'duration_ms',
       'acousticness', 'danceability', 'energy', 'instrumentalness',
       'liveness', 'loudness', 'speechiness', 'valence', 'key', 'mode',
       'tempo', 'time_signature']


# empty list to append to
features_data = []

# counter variable
# increase with each loop
count = 9


for id_list in track_id_split[8:10]:
    
    import spotipy
    import spotipy.util as util
    import spotipy.oauth2 as oauth2
    
    ### SET CREDENTIALS ###
    credentials = oauth2.SpotifyClientCredentials(client_id=client_id, 
                                                  client_secret=client_secret)
    
    token = credentials.get_access_token()
    sp = spotipy.Spotify(auth=token)
    
    sp.trace = False
    sp.trace_out = False
    
    print('Credentials set')
    print('This is loop number {}'.format(count))
    
    ### extract features
    for track in id_list:
          
        try:
            # extract features
            features = sp.audio_features(track)

            # append to features_data
            features_data.append(features[0])

            # print(features_data[-1]['uri'], features_data[-1]['energy'], len(features_data))

        except (TypeError, AttributeError):
            print('error: ',track)

        sleep(1) # don't be a jerk
    
    ### CREATE DF ###
    print('creating DF')
    features_df = pd.DataFrame(features_data, columns=cols)
    
    # rename columns
    features_df.columns = new_cols
    
    # save the df
    f_name = 'rc_features_df_{}.csv'.format(count)
    
    features_df.to_csv(f_name, index=False)
    
    print(len(features_df['id']))
    print('new df created')
    
    # add 1 too the loop counter
    count += 1
```
